```python
from google.colab import drive
import os

# Mount Google Drive
drive.mount('/content/drive')

# Define the root directory based on your Google Drive path
root_dir = '/content/drive/My Drive/UAS_DL/DATASET'

# Define the dataset folders to search
dataset_folders = [
    'After_Split/Dataset_White',
    'CROP_AFTER_SPLIT_MIX'
]

# Function to count jpg images in a directory
def count_images_in_directory(directory):
    return sum([len(files) for r, d, files in os.walk(directory) if any(file.endswith('.jpg') for file in files)])

# Traverse the dataset folders and count images
image_counts = {}

for dataset in dataset_folders:
    dataset_path = os.path.join(root_dir, dataset)

    if 'CROP_AFTER_SPLIT_MIX' in dataset:
        subfolders = ['val', 'train', 'test']
        for subfolder in subfolders:
            subfolder_path = os.path.join(dataset_path, subfolder)
            class_folders = [f.path for f in os.scandir(subfolder_path) if f.is_dir()]
            for class_folder in class_folders:
                class_name = os.path.basename(class_folder)
                image_count = count_images_in_directory(class_folder)
                image_counts[f'{dataset}/{subfolder}/{class_name}'] = image_count
    else:
        class_folders = [f.path for f in os.scandir(dataset_path) if f.is_dir()]
        for class_folder in class_folders:
            class_name = os.path.basename(class_folder)
            image_count = count_images_in_directory(class_folder)
            image_counts[f'{dataset}/{class_name}'] = image_count

# Display the counts
for path, count in image_counts.items():
    print(f'{path}: {count} images')
```
